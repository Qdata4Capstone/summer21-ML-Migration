```python
rejectshare_grp = reject_share.groupby(['year','iso'])
```

```python
distance_grp = distance_dataset.groupby(['iso_o'])
```

```python
no_distance_countries = []
no_application_countries = []
no_rejectshare_countries = []
```

```python
for key, item in grp:
    try:
        apprehensions = app_grp.get_group(key)['apprehensions'].item()
    except:
        apprehensions = 0
    try:
        rejectshare_applications = rejectshare_grp.get_group(key)['rejectshare'].item()
    except:
        if key[0]>=2000:
            no_rejectshare_countries.append(key)
        rejectshare_applications = 0
    try:
        asylum_applications = us_grp.get_group(key)['US_asylum_applications'].item()
        print(asylum_applications)
    except:
        if key[0]>=2007:
            no_application_countries.append(key)
        asylum_applications = 0
    try:
        mindist = distance_grp.get_group(key[1])['mindist'].item() 
        popdist = distance_grp.get_group(key[1])['dist'].item()
    except:
        mindist=0
        popdist=0
        no_distance_countries.append(key[1])
        print("Not found country ISO: ",key[1])
    print(key,apprehensions,rejectshare_applications,asylum_applications,mindist,popdist)
    original_dataset['US_apprehensions'].loc[(original_dataset['year']==key[0]) & (original_dataset['ISO3']==key[1])] = apprehensions
    original_dataset['US_asylum_applications'].loc[(original_dataset['year']==key[0]) & (original_dataset['ISO3']==key[1])] = asylum_applications
    original_dataset['perc_rejected_asylum'].loc[(original_dataset['year']==key[0]) & (original_dataset['ISO3']==key[1])] = rejectshare_applications
    original_dataset['min_distance'].loc[(original_dataset['ISO3']==key[1])] = mindist
    original_dataset['pop_distance'].loc[(original_dataset['ISO3']==key[1])] = popdist
    
```
